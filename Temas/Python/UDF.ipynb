```python
import sqlite3
import pandas as pd
```

```python
square = lambda n : n*n
```

# Ejercicio 1

```python
conn = sqlite3.connect("./../../northwind.db")

conn.create_function("square",1,square)

cursor = conn.cursor()
cursor.execute('''
    SELECT * FROM Products
    '''
)

Results = cursor.fetchall()
df_products = pd.DataFrame(Results)

conn.commit()

cursor.close()
conn.close()

print(df_products)
```

# Ejercicio 2

```python
with sqlite3.connect("./../../northwind.db") as conn:
    conn.create_function("square",1,square)
    cursor = conn.cursor()
    cursor.execute("SELECT *,square(price) as Precio_cuadrado FROM Products where price > 0")
    results = cursor.fetchall()
    results_df = pd.DataFrame(results)

print(results_df)
```
